| **Status Code**       | **Description**                              | **Further Explanation**                                                                 |
|------------------------|----------------------------------------------|-----------------------------------------------------------------------------------------|
| **200 OK**             | Standard success                            | A GET or POST succeeded.                                                               |
| **201 Created**        | Resource created                            | After a POST that creates something.                                                   |
| **204 No Content**     | Success, but no data returned               | After a DELETE request.                                                                |
| **400 Bad Request**    | Malformed or invalid request                | Something is wrong with the request from the client.                                    |
| **401 Unauthorized**   | No/invalid authentication                   | The client is not authenticated.                                                       |
| **403 Forbidden**      | Authenticated, but no permission            | The client is authenticated but does not have permission to access the resource.        |
| **404 Not Found**      | Resource doesn't exist                      | The requested resource could not be found.                                             |
| **500 Internal Server Error** | Generic failure                     | Something went wrong on the server side.                                               |
| **502 Bad Gateway**    | Gateway (like Nginx) failed to reach backend | The gateway server failed to get a valid response from the backend server.             |
| **503 Service Unavailable** | Server is down or overloaded           | The server is temporarily unable to handle the request (e.g., maintenance or overload). |
| **Redirection**        | Further action needs to be taken            | For example, the client needs to follow a redirect.                                     |
| **Client Error**       | Something is wrong with the request         | Indicates issues with the clientâ€™s request.                                            |
| **Server Error**       | Something went wrong on the server side     | Indicates issues on the server side.                                                   |
